# BODAQS Persisted Entity Scope Example

This notebook demonstrates the explicit persisted-scope workflow:

1. Use a live entity selector with autosave enabled.
2. Build a separate selector-compatible persisted scope handle.
3. Feed that persisted handle into a downstream widget.

```python
from IPython.display import display
from bodaqs_analysis.widgets.session_selector import make_session_selector

sel = make_session_selector(artifacts_dir='artifacts', select_first_by_default=True)
display(sel['ui'])
```

```python
from IPython.display import display
from bodaqs_analysis.widgets.entity_scope_store import make_persisted_entity_scope_handle
from bodaqs_analysis.widgets.session_selector import attach_refresh
from bodaqs_analysis.widgets.signal_histogram_widget import make_signal_histogram_rebuilder

persisted = make_persisted_entity_scope_handle(artifacts_dir='artifacts')
display(persisted['ui'])

hist = make_signal_histogram_rebuilder(sel=persisted, registry_policy='union')
display(hist['out'])

refresh = attach_refresh(persisted, [hist['rebuild']])
```
